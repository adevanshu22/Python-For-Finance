```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['PG', 'MSFT', 'F', 'GE']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo', start = '1995-1-1')['Adj Close']
```

```python
mydata.info()
```

```python
mydata.head()
```

```python
mydata.tail()
```

```python
mydata.iloc[0:5]
```

```python
#Normalize values to 100
((mydata/mydata.iloc[0]) * 100).plot(figsize = (15, 5))

# Compare behavior of all different stocks as if all are starting from same value i.e. 100
```

```python
# Note that in the above since the timehorizon is very large i.e 25 years, Microsoft has given huge returns. But if we shorten the time period, we can observe clear differences
((mydata.head(5000)/mydata.iloc[0]) * 100).plot(figsize = (15, 5))

```

```python
mydata.plot(figsize = (15, 5))
```

## Calculating simple return of these stocks

```python
returns = (mydata/mydata.shift(1)) - 1
returns.head()
```

```python
weights = np.array([0.25, 0.25, 0.25, 0.25])
```

```python
weights.shape
```

```python
np.dot(returns, weights)
```

```python
annual_returns = returns.mean()*250
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights), 5) * 100) + '%'
print(pfolio_1)
```

```python
weights_2 = np.array([0.4, 0.4, 0.15, 0.05])
```

```python
pfolio_2 = str(round(np.dot(annual_returns, weights_2), 5) * 100) + '%'
print(pfolio_2)
```
